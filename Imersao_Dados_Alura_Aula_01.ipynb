```
# Importação de bibliotecas
import pandas as pd
```

```
# Link para o arquivo de dados
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'
```

```
#Importação do arquivo de dados
dados = pd.read_csv(url_dados,compression='zip')
```

# Tratamento do dataframe

```
#Desafio 06: Renomear as colunas tirando o hífen.
new_col_name = {}
add_to_dict = lambda key : key.replace('-','')

for i in dados.columns:
  new_col_name[i] = add_to_dict(i)

dados.rename(new_col_name, axis = 'columns')
```

# Exploração dos dados

```
#Desafio 04: Quantos tipos de drogas foram investigados.
dados['droga'].nunique()
```

```
#Desafio 07: Deixar os gráficos bonitões.
ax = dados['tempo'].value_counts().plot.bar()
ax.set_title('Quantidade de Ocorrência X Tempo',fontsize = 16)
ax.set_ylabel('Quantidade',fontsize = 12)
ax.set_xlabel('Tempo',fontsize = 12)
```

```
#Desafio 01: Investigar por que a classe tratamento é tão desbalanceada?
# Analisando a quantidade de ocorrências para cada item da série 'tempo',
# agrupada por tipo de dose e de tratamento, infere-se que um mesmo grupo
# de controle pode ser utilizado para vários tipos de droga.

#Desafio 03: Proporção das classes tratamento
# Para analisar a classe tratamento é necessário agrupar a informação de
# acordo com o tipo de dose e intervalo em que são ministradas.

dados.groupby(['dose','tratamento'])[('tempo')].value_counts()
```
